# Exploratory Data Analysis

## Gender

```python
# plotting bar chart for gender and its count
gb= df.groupby(['Gender'], as_index= False) ['Order ID'].count().sort_values(by= 'Gender', ascending= False)
ax = plt.pie(gb['Order ID'], labels=gb['Gender'], autopct='%1.1f%%', startangle=140, colors=['red', 'blue'])
plt.show()
```

```python
# now we will check that whether male or female who has spent more amount.
gb= df.groupby(['Gender'], as_index= False) ['Amount'].sum().sort_values(by= 'Amount', ascending= False)
ax= sns.barplot(x= 'Gender', y='Amount', data= gb, hue= 'Gender')
sns.set(rc={'figure.figsize':(5,5)})
plt.title('Gender Vs Amount')

```

**from above graphs we can conclude that Female customers are more as compared to male customers also they are spending more money or we can say that
they have high purchasing power**

## Age Group

```python
# plotting count plot to check count of gender in each age group
ax = sns.countplot(x='Age Group', data = df, hue= 'Gender')
plt.title('Age Group')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# now we will check that which age group has spent more amount.
gb= df.groupby(['Age Group'], as_index= False) ['Amount'].sum().sort_values(by= 'Amount', ascending= False)
ax= sns.barplot(x= 'Age Group', y='Amount', data= gb, hue = 'Age Group')
plt.title('Age Group Vs Amount')
```

**from above graphs we can conclude that most of the customers are of middle age group and also the females of middle age group are 
more as compared to that of male**

## Month

```python
# plotting countplot to know orders per month
ax = sns.countplot(x='Month', data = df, hue= 'Month',order= df['Month'].value_counts().index)
plt.title('Month')
for bars in ax.containers:
    ax.bar_label(bars)
sns.set(rc={'figure.figsize':(10,5)})
```

```python
# now we will check that which age group has spent more amount.
gb= df.groupby(['Month'], as_index= False) ['Amount'].sum().sort_values(by= 'Amount', ascending= False)
ax= sns.barplot(x= 'Month', y='Amount', data= gb, hue = 'Month')
plt.title('Month Vs Amount')
sns.set(rc={'figure.figsize':(20,5)})
```

**from above graphs we can conclude that the most orders are recieved in month march , february and january but the most revenue was generated in month march, february and april**

## Status

```python
# plotting countplot to know status of orders
gb= df.groupby(['Status'], as_index= False) ['Order ID'].count().sort_values(by= 'Status', ascending= False)
ax = plt.pie(gb['Order ID'], labels=gb['Status'], autopct='%1.1f%%', startangle=140, colors=['red', 'orange', 'green', 'blue'])
plt.show()
```

**from above graph we can see that 92.3% of total orders were delievered**


## Category

```python
# plotting countplot to know category with most sales
ax = sns.countplot(x='Category', data = df, hue= 'Category',order= df['Category'].value_counts().index)
plt.title('Category')
for bars in ax.containers:
    ax.bar_label(bars)
sns.set(rc={'figure.figsize':(20,5)})
```

```python
# now we will check that which category has generated high revenue
gb= df.groupby(['Category'], as_index= False) ['Amount'].sum().sort_values(by= 'Amount', ascending= False)
ax= sns.barplot(x= 'Category', y='Amount', data= gb, hue = 'Category')
plt.title('Category Vs Amount')
sns.set(rc={'figure.figsize':(20,5)})
```

**from above graphs we can conclude that most sold products are of category Set**

## Size


```python
# plotting countplot to know Size with most sales
gb= df.groupby(['Size'], as_index= False) ['Order ID'].count().sort_values(by= 'Order ID', ascending= False)
ax= sns.barplot(x= 'Size', y='Order ID', data= gb, hue = 'Size')
plt.title('Size')
sns.set(rc={'figure.figsize':(20,5)})
```

**from above graph we can say that most sold sizes are medium, large and XL respectively**

## State 


```python
# plotting countplot to know States with high sales
gb= df.groupby(['ship-state'], as_index= False) ['Order ID'].count().sort_values(by= 'Order ID', ascending= False).head(12)
ax= sns.barplot(x= 'ship-state', y='Order ID', data= gb, hue = 'ship-state')
plt.title('State')
sns.set(rc={'figure.figsize':(20,5)})
```

```python
gb= df.groupby(['ship-state'], as_index= False) ['Amount'].count().sort_values(by= 'Amount', ascending= False).head(12)
ax= sns.barplot(x= 'ship-state', y='Amount', data= gb, hue = 'ship-state')
plt.title('State Vs Amount')
sns.set(rc={'figure.figsize':(20,5)})
```

**from above graphs we can conclude that Maharshta, Karnataka, Uttar Pradesh are top 3 states from which most of the orders are recieved and most of the revenue is generated**

## City

```python
# plotting countplot to know cities with high sales
gb= df.groupby(['ship-city'], as_index= False) ['Order ID'].count().sort_values(by= 'Order ID', ascending= False).head(12)
ax= sns.lineplot(x= 'ship-city', y='Order ID', data= gb)
plt.title('City')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
gb= df.groupby(['ship-city'], as_index= False) ['Amount'].count().sort_values(by= 'Amount', ascending= False).head(12)
ax= sns.lineplot(x= 'ship-city', y= 'Amount',data= gb)
plt.title('City Vs Amount')
sns.set(rc={'figure.figsize':(20,5)})
```

**from above graphs we can conclude that Bengaluru, Hyderabad, New Delhi are the top 3 cities with high no. of orders and high revenue**
